```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
companies = pd.read_csv('../data/01 - cleaned/companies_cleaned.csv')
```

Se analiza cuántas empresas tienen review.

```python
(
    (companies.rating != 0)
    .replace({True: "Con reviews", False: "Sin reviews"})
    .value_counts(normalize=True)
)
```

El 74% de las empresas que no tienen review, ningún cliente ha hecho una apreciación del vínculo.

Se proyecta la distribución de rating y reviews.

```python
empresas_con_rating = companies.query('reviews != 0')
plt.scatter(empresas_con_rating.reviews, empresas_con_rating.rating)
plt.xlabel("Reviews")
plt.ylabel("Rating")
plt.title("Rating vs Reviews")
plt.show()
```

```python
companies.query('rating != 0').rating.astype(float).plot.hist()
plt.title("Histograma de rating")
plt.xlabel("Rating")
plt.show()
```

Debido a que Clutch certifica las reviews, es muy probable que la misma este sesgada a las reviews positivas.

Esto último se ve claramente reflejado en el histograma anterior.
